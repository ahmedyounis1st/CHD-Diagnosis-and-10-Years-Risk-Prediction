```python
def creat_prediction():
    global gender2, age_2, education2, current_smoker2, cigs_per_day2, bp_meds2
    global prevalent_stroke2, prevalent_hyp2, diabetes2, tot_chol2, sys_bp2, dia_bp2, bmi2, heart_rate2, glucose2
    global top, pred_label
    if top != None:
        top.destroy()
    top = Tk()
    top.title("Coronary Heart Disease Risk")
    top.minsize(755,512)
    L = Label(top, text= "Test risk for 10 years!")
    L.pack(pady = 10)
    frame3 = LabelFrame(top, text = "Initialization ", padx = 50,pady = 10)
    frame3.pack(pady = 10)
    
    gender1  = Label(frame3, text = "Gender (0,1): ", anchor = W)
    gender1.grid(row = 0, column = 0 , pady = 5)
    gender2 = Entry(frame3)
    gender2.grid(row = 0, column = 1, pady = 5)
    
    age_1  = Label(frame3, text = "Age: ", anchor = W)
    age_1.grid(row = 1, column = 0 , pady = 5)
    age_2 = Entry(frame3)
    age_2.grid(row = 1, column = 1, pady = 5)
    
    education1  = Label(frame3, text = "Education: ", anchor = W)
    education1.grid(row = 2, column = 0 , pady = 5)
    education2 = Entry(frame3)
    education2.grid(row = 2, column = 1, pady = 5)
    
    current_smoker1  = Label(frame3, text = "Current smoker (0,1): ", anchor = W)
    current_smoker1.grid(row = 3, column = 0 , pady = 5)
    current_smoker2 = Entry(frame3)
    current_smoker2.grid(row = 3, column = 1, pady = 5)
    
    cigs_per_day1  = Label(frame3, text = "Number of Cigs per day: ", anchor = W)
    cigs_per_day1.grid(row = 4, column = 0 , pady = 5)
    cigs_per_day2 = Entry(frame3)
    cigs_per_day2.grid(row = 4, column = 1, pady = 5)
    
    bp_meds1  = Label(frame3, text = "Blood pressure medicine (0,1): ", anchor = W)
    bp_meds1.grid(row = 5, column = 0 , pady = 5)
    bp_meds2 = Entry(frame3)
    bp_meds2.grid(row = 5, column = 1, pady = 5)
    
    prevalent_stroke1  = Label(frame3, text = "Prevalent stroke (0,1): ", anchor = W)
    prevalent_stroke1.grid(row = 6, column = 0 , pady = 5)
    prevalent_stroke2 = Entry(frame3)
    prevalent_stroke2.grid(row = 6, column = 1, pady = 5)
    
    prevalent_hyp1  = Label(frame3, text = "Prevalent hypertension (0,1): ", anchor = W)
    prevalent_hyp1.grid(row = 7, column = 0 , pady = 5)
    prevalent_hyp2 = Entry(frame3)
    prevalent_hyp2.grid(row = 7, column = 1, pady = 5)
    
    diabetes1  = Label(frame3, text = "Diabetes (0,1): ", anchor = W)
    diabetes1.grid(row = 8, column = 0 , pady = 5)
    diabetes2 = Entry(frame3)
    diabetes2.grid(row = 8, column = 1, pady = 5)
    
    tot_chol1  = Label(frame3, text = "Cholesterol: ", anchor = W)
    tot_chol1.grid(row = 9, column = 0 , pady = 5)
    tot_chol2 = Entry(frame3)
    tot_chol2.grid(row = 9, column = 1, pady = 5)
    
    sys_bp1  = Label(frame3, text = "Systolic blood pressure: ", anchor = W)
    sys_bp1.grid(row = 10, column = 0 , pady = 5)
    sys_bp2 = Entry(frame3)
    sys_bp2.grid(row = 10, column = 1, pady = 5)
    
    dia_bp1  = Label(frame3, text = "Diastolic blood pressure: ", anchor = W)
    dia_bp1.grid(row = 11, column = 0 , pady = 5)
    dia_bp2 = Entry(frame3)
    dia_bp2.grid(row = 11, column = 1, pady = 5)
    
    bmi1  = Label(frame3, text = "Body mass index: ", anchor = W)
    bmi1.grid(row = 12, column = 0 , pady = 5)
    bmi2 = Entry(frame3)
    bmi2.grid(row = 12, column = 1, pady = 5)
    
    heart_rate1  = Label(frame3, text = "Heart rate: ", anchor = W)
    heart_rate1.grid(row = 13, column = 0 , pady = 5)
    heart_rate2 = Entry(frame3)
    heart_rate2.grid(row = 13, column = 1, pady = 5)
    
    glucose1  = Label(frame3, text = "Glucose: ", anchor = W)
    glucose1.grid(row = 14, column = 0 , pady = 5)
    glucose2 = Entry(frame3)
    glucose2.grid(row = 14, column = 1, pady = 5)

         
    btn2 = Button(top , text = 'Predict', command = predict, padx = 20, pady =10, bg='brown', fg='white')
    btn2.pack_forget()
    btn2.pack(padx = 30, pady= 20)
    pred_label = Label(top, text = "Prediction: ")
    
    top.mainloop()
```

```python
def predict():
    global gender2, age_2, education2, current_smoker2, cigs_per_day2, bp_meds2
    global prevalent_stroke2, prevalent_hyp2, diabetes2, tot_chol2, sys_bp2, dia_bp2, bmi2, heart_rate2, glucose2
    global pred_label
    pred_input = []
    pred_input.append(int(gender2.get()))
    pred_input.append(int(age_2.get()))
    pred_input.append(int(education2.get()))
    pred_input.append(int(current_smoker2.get()))
    pred_input.append(float(cigs_per_day2.get()))
    pred_input.append(int(bp_meds2.get()))
    pred_input.append(int(prevalent_stroke2.get()))
    pred_input.append(int(prevalent_hyp2.get()))
    pred_input.append(int(diabetes2.get()))
    pred_input.append(float(tot_chol2.get()))
    pred_input.append(float(sys_bp2.get()))
    pred_input.append(float(dia_bp2.get()))
    pred_input.append(float(bmi2.get()))
    pred_input.append(float(heart_rate2.get()))
    pred_input.append(float(glucose2.get()))
    
    X = np.array(pred_input).reshape(1, -1)
    X[:,index] = pred_scaler.transform(X[:,index])
    y = pred_model.predict(X)
    
    pred_label = Label(top, text = f"Prediction: {y}")
    pred_label.pack_forget()
    pred_label.pack()
```

```python
def diagnosis():
    global diag_label
    diag_input = []
    diag_input.append(float(sbp2.get()))
    diag_input.append(float(tobacco2.get()))
    diag_input.append(float(ldl2.get()))
    diag_input.append(float(adiposity2.get()))
    diag_input.append(int(famhist2.get()))
    diag_input.append(float(typea2.get()))
    diag_input.append(float(obesity2.get()))
    diag_input.append(float(alcohol2.get()))
    diag_input.append(int(age2.get()))
    
    X = np.array(diag_input).reshape(1, -1)
    X[:,Index] = diag_scaler.transform(X[:,Index])
    y = diagnosis_acc_model.predict(X)
    
    diag_label = Label(frame1, text = f"Diagnosis: {y}")
    diag_label.pack_forget()
    diag_label.pack()
    if (y ==  0):
        creat_prediction()
```
